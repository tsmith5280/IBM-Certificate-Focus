```python
# Task 2: Mark the success/failed launches for each site on the map
# Create a map centered on the US
site_map = folium.Map(location=[28.562302, -80.577356], zoom_start=4)

# Loop through each row in the SpaceX dataframe
for index, row in spacex_df.iterrows():
    # Choose color based on success (1) or failure (0)
    color = 'green' if row['class'] == 1 else 'red'
    
    # Add marker for each launch
    folium.CircleMarker(
        location=[row['Lat'], row['Long']],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=row['Launch Site']
    ).add_to(site_map)

# Display the map
site_map
```

Most SpaceX launches occur in Florida (KSC/CCAFS), which is close to the equator to benefit from Earth's rotational speed.

All launch sites are near the ocean. This minimizes risk to populated areas and allows rocket parts to safely fall into the sea.

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordinates (replace with actual values you get from MousePosition)
launch_site_lat, launch_site_lon = 28.563197, -80.576820
coastline_lat, coastline_lon = 28.56367, -80.57163

# Calculate distance
distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Add distance marker to the map
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance),
    )
).add_to(site_map)

# Draw a line
line = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    weight=2
).add_to(site_map)
```
